###4. Program to **find ROOTS of QUADRATIC EQUATIONS**.

```python
print( 'Name: Kavyam Patel')
print('Class and sec: X A')
print('')

# Solve the quadratic equation ax**2 + bx + c = 0
# import complex math module

import cmath
a = int(input("enter value of a for b**2-4ac = "))
b = int(input("enter value of b for b**2-4ac= "))
c = int(input("enter value of c for b**2-4ac= "))

# calculate the discriminant
d = (b**2) - (4*a*c)
print("Discriminant is",d)
        
# find two solutions
root1 = (-b-cmath.sqrt(d))/(2*a)
root2 = (-b+cmath.sqrt(d))/(2*a)
print('The solution are ',(root1,root2))
```

###10. Print **STAR PATTERN**

```python
print('')
print('Name: Kavyam Patel')
print('Class and Sec: X A')

# Print * Pattern

rows = 6

# Outer Loop

for i in range(rows):
    for j in range(i):
        print(i, end='')
    print('')

# Print Reversed pattern

rows = 5
b = 0

# Reverse For Loop from 5 to 0

for i in range(rows, 0, -1):
    b +=1
    for j in range(1, i+1):
        print(b, end='')
    print('\r')
print('\n')
```

###11. Program to **find if a Number is PRIME or NOT**.

```python
print('')
print('Name: Kavyam Patel')
print('Class and Sec: X A')

# Python Progran to check if given number is prime or not

num = float(input('Enter a Whole Number: '))

num = int(num)

if num==1 or num<=0:
    print('\nInvalid Number!')
    print(f'{num} is niether a Prime nor a Composite number.')

elif num<=0:
    print('\nInvalid Number')
    print('Please Enter a Positive Number.')

else:
    for i in range(2, num):
        if (num%i)==0:
            print(num, ' is not a prime number.')
            break
        else:
            print(num, ' is a prime number.')
            break
print('')
```

###13. **MENU** driven program using **WHILE Loop**.

```python
print('\nName: Kavyam Patel')
print('Class and Sec: X A')

while True:
    print('\nCACULATE AREA')
    print('\n1. Circle')
    print('2. Rectangle')
    print('3. Square')
    print('4. Exit')
    print('')
    Choice = float(input('Enter your choice: '))
    print('')
    if Choice == 1:
        Radius = float(input('Enter the RADIUS: '))
        Area = 3.14 * Radius * Radius
        Area = round(Area, 2)
        print('Area of Circle =', Area)
        break
    
    elif Choice == 2:
        Lenght = float(input('Enter the Lenght: '))
        Breadth = float(input('Enter the Breadth: '))
        Area = Lenght * Breadth
        Area = round(Area, 2)
        print('Area of Rectangle = ', Area)
        break

    elif Choice == 3:
        Side = float(input('Enter lenth of the Side: '))
        Area = Side * Side
        Area = round(Side, 2)
        print('Area of Square = ', Area)
        break

    elif Choice == 4:
        print('Bye !!!')
        break

    else:
        print('Oops! Incorrect Choice')

print('')
```

###14. Program to **find SUM of DIGITS of a Number using WHILE Loop**

```python
print('\nName: Kavyam Patel')
print('Class and Sec: X A')

n = int(input('Enter Number: '))
sum = 0
while(n != 0):
    sum = sum + (n % 10)
    n = n // 10

print('\nSum of the digits of the entered number is: ', sum)
print('')
```
